```python
build = soup.find_all('a', attrs= {'target':'_blank'})
print(build)
```

```python
b = soup.find_all('span', attrs = {'itemprop':'addressLocality'})
places = [i.text for i in b]
print(places)
print(len(places))
```

```python
c = soup.find_all('div', attrs={'data-type':'price-link'})
d = [i.text.strip() for i in c]
print(d)
print(len(d))
```

```python
price = []
for i in c:
    i = i.text.strip()
    if 'Cr' in i:
        i = i.replace(' Cr', ' ')
    else:
        i = i.replace(' L', ' ')
    price.append(i)
print(price)
```

```python
d = soup.find_all('td', attrs ={'class':'size'})
area = [i.text.strip() for i in d]
print(area)
print(len(area))
```

```python
property_types =['Apartment', 'Builder floor', 'Villa', 'Resdential Plot', 'Independent House', 'Independent Floor', 'Studio Apartment']

f = soup.find_all('div', attrs = {'class':'title-line'})
Type = []
for i in f:
    i = i.text
    for p_type in property_types:
        if p_type in i :
            Type.append(p_type)
print(Type)
print(len(Type))
```

```python
f = soup.find_all('div',attrs = {'class':'title-line'})
Rooms = []
for i in f:
    i = i.span.text
    i = i.replace('ResidentialPlot', '0').replace(' ', '')
    Rooms.append(i)
print(Rooms)
```

```python
g = soup.find_all('td', attrs = {'class':'val'})
Rtm = []
for i in g:
    i = i.text
    Rtm.append(i)
print(Rtm)
```

```python
g = soup.find_all('ul', attrs = {'class':'listing-details'})
details = [i.text for i in g]
print(details)
print(len(details))
```

```python
import re # pattern matching

d = soup.find_all('ul', attrs = {'class':'listing-details'})
bathrooms = []
facing = []
for i in d:
    i = i.text
    bathroom_count = re.findall(r'(\d+) Bathrooms', i) #d - digits
    if bathroom_count:
        bathrooms.append(bathroom_count[0])
    else:
        bathrooms.append(0)
    facing_direction = re.findall(r'(North|South|East|West|NorthEast|NortWest|SouhEast|SouthWest)facing', i)

    if facing_direction:
        facing.append(facing_direction[0])
    else:
        print('None')

print(facing)
print(bathrooms)
```
